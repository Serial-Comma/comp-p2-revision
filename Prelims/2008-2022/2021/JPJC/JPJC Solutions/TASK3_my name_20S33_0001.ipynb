```python
#Task 3.1
import datetime
this_year = datetime.datetime.today().year
now = datetime.datetime.today()

class Member(object): 
    def __init__(self, memberID, firstName, surname, contact):
        
        self.__memberID = memberID
        self.__first_name = firstName
        self.__surname = surname
        self.__contact_number = contact
        self.__last_visit = str(now)[0:10]
        self.__memberType = None

    def showMember(self):
        print()
        print("Member ID:", self.__memberID)
        print("First name:", self.__first_name)
        print("Surname:", self.__surname)
        print("Contact:", self.__contact_number)
        print("Last Visit:", self.__last_visit)

    def isActive(self):
        now = datetime.datetime.today()
        lastTime = datetime.datetime.strptime(self.__last_visit,"%Y-%m-%d")
        if (now - lastTime).days < 31:
            return True
        return False

    def getMemberID(self):
        return self.__memberID
    def getFirstName(self):
        return self.__first_name
    def getSurname(self):
        return self.__surname
    def getContactNo(self):
        return self.__contact_number
    def getLastVisit(self):
        return self.__last_visit
    def setMemberType(self, memType):
        self.__memberType = memType
    def getMemberType(self):
        return self.__memberType

class NormalMember(Member):
    def __init__(self, memberID, firstName, surname, contact):
        super().__init__(memberID, firstName, surname, contact)
        self.__stored_value = float(0)
        self.setMemberType("normal")

    def showMember(self):
        super().showMember()
        print("Stored value: {:.2f}".format(self.__stored_value))
        print("Member Type:", self.getMemberType())
    def getStoredValue(self):
        return "${:0.2f}".format(self.__stored_value)
    def topup(self, amt):   #not stated in question
        self.__stored_value = self.__stored_value+float(amt)
    def deduct(self, amt):   #not stated in question
        self.__stored_value = self.__stored_value-float(amt)

class AnnualMember(Member):
    def __init__(self, memberID, firstName, surname, contact):
        super().__init__(memberID, firstName, surname, contact)
        self.__annual_fee = int(500)
        self.__date_register = str(now)[0:10]
        self.setMemberType("annual")
        
    def showMember(self):
        super().showMember()
        print("Annual fee:", self.__annual_fee)
        print("Date registered:", self.__date_register)
        print("Member Type:", self.getMemberType())
    def getAnnualFee(self):
        return "{}".format(self.__annual_fee)
    def setAnnualFee(self, amt):
        self.__annual_fee = int(amt)
    def getDateRegister(self):
        return self.__date_register
    def setDateRegister(self):
        self.__date_register = str(now)[0:10]

#Task 3.2
with open('members.txt','r') as file:
    memList = []
    for line in file:
        temp = line.strip().split(',')
        if temp[-1] == "normal":
            memList.append(NormalMember(temp[0],temp[1],temp[2],temp[3]))
        else:
            memList.append(AnnualMember(temp[0],temp[1],temp[2],temp[3]))
for each in memList:
    each.showMember()

#Task 3.3
import sqlite3
db = sqlite3.connect("jpgym.db")
db.execute("CREATE TABLE IF NOT EXISTS 'Member' ('MemberID' TEXT PRIMARY KEY, 'FirstName' TEXT,\
'Surname' TEXT, 'ContactNo' TEXT, 'LastVisit' TEXT, 'MemberType' TEXT);")
for each in memList:
    db.execute("INSERT INTO Member(MemberID,FirstName,Surname,ContactNo,LastVisit,MemberType) VALUES\
(?,?,?,?,?,?)",(each.getMemberID(), each.getFirstName(), each.getSurname(), each.getContactNo(), each.getLastVisit(), each.getMemberType()))
db.commit()

#Task 3.4
row = db.execute("SELECT FirstName, Surname, ContactNo FROM Member WHERE MemberType='normal' ORDER BY FirstName")
for cur in row:
    print(cur)
db.close()

```
