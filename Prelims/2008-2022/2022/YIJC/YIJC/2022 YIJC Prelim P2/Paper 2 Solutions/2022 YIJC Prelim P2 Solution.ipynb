```python
# Haversine function provided

from math import *

def Haversine(coord1,coord2):

    long1, lat1 = coord1
    long2, lat2 = coord2

    R = 6371000                               # radius of Earth in meters
    phi_1 = radians(lat1)
    phi_2 = radians(lat2)

    delta_phi = radians(lat2-lat1)
    delta_lambda = radians(long2-long1)

    a = sin(delta_phi/2.0)**2 + cos(phi_1) * cos(phi_2) * sin(delta_lambda/2.0)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))

    return R * c/1000.0


pt1 = [103.7423112,1.33315262]
pt2 = [103.74959,1.349034109]

Haversine(pt1, pt2)
```

```python
#Task 2.1

from random import *

def fence():
    lst = []
    for y in range(22):            
        for x in range(32):
            if y==0 or y==21:
                lst.append([x,y])
            elif x==0 or x==31:
                lst.append([x,y])
    return lst

field = fence()

snake = [15,10]

def food_loc(snake_loc, region):
    x, y = 0, 0
    while [x,y] in region or [x,y]==snake_loc:
        x = randint(1,31)
        y = randint(1,21)
    return [x,y]

food = food_loc(snake, field)

def display(s_loc, f_loc, region):
    for y in range(22):
        string = ''
        for x in range(32):
            if [x,y] == s_loc:
                string += 'O'
            elif [x,y] == f_loc:
                string += 'F'
            elif [x,y] in region:
                string += 'X'
            else:
                string += '.'
        print(string)

display(snake, food, field)
```

```python
#Task 3.2
                
class BSTree():
    def __init__(self):
        self.ThisTree=[] 
        self.Root=None
        self.NextFree=0
        
    def AddItem(self,NewTreeItem):
        if self.Root==None:
            New=Node(NewTreeItem)
            self.ThisTree.append(New)
            self.Root=self.NextFree
            self.NextFree+=1       
        else:
            Current=self.Root
            LastMove='X'

            while Current !=None:
                Previous=Current
                if NewTreeItem<self.ThisTree[Current].GetData():
                    LastMove='L'
                    Current=self.ThisTree[Current].GetLeft()
                else:
                    LastMove='R'
                    Current=self.ThisTree[Current].GetRight()

            if LastMove=='R':
                self.ThisTree[Previous].SetRight(self.NextFree)
            else:
                self.ThisTree[Previous].SetLeft(self.NextFree)

            New=Node(NewTreeItem) 
            self.ThisTree.append(New)
            self.NextFree+=1
```
