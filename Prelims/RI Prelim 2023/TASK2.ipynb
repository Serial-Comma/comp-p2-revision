```python
#Task 2.1


hellonum = 0

def proccess_input(input_text):
    input = input_text
    global hellonum
    if "hello" in input.lower():
        if hellonum != 0:
            return "Hello again, welcome back!"
        else:
            hellonum+=1
            return "Hi, how are you?"
    elif "thanks" in input.lower() or "thank you" in input.lower():
        return "You are most welcome."
    elif input[0] == "I"and "you" in input[1:].lower():
        youindex = input.lower().index("you")
        # youindex = youindex-1 
        xyz = input[2:youindex-1]
        return f"You {xyz} me? I really {xyz} you too."
    else:
        return "Sorry, I do not understand..."
    
# proccess_input("Well, hello there!")  
# proccess_input("This is a test")
# proccess_input("I am fine, thank you.")
# proccess_input("Wait, I have to go...")
# proccess_input("HELLO! I have returned.")
# proccess_input("I missed you so much!")
# proccess_input("I thank you... hello?")
# proccess_input("You and I think alike, you know...")




def main():
    import socket

    PORT = 12345
    SERVER = "127.0.0.1"
    ADDR = (SERVER,PORT)

    server = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
    server.bind(ADDR)
    def handle_client(conn,addr):
        connected = True
        while connected == True:
            msg = conn.recv(2048).decode()
            if msg:
                if "exit" in msg.lower():
                    connected = False
                    print("Client disconnected!")
                    break
                print(f"Client: {msg}")
                response = proccess_input(msg) + "\n"
                conn.sendall(response.encode())
                print(f"Server: {response}")

                    
        conn.close()


    def start():
        server.listen()
        while True:
            conn,addr = server.accept()
            handle_client(conn,addr)

    start()



main()
```
