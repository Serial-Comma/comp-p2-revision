```python
#Task 1.5



def insertionsort(seq):
    arr =seq 

    sortlist = []
    for i in range(len(arr)):
        insert(int(arr[i]),sortlist)
    return sortlist

print(insertionsort(lst))
```

```python
#Task 1.6

def bubblesort(seq):
    arr = seq
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i] < arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr

bubblesort(lst)
```

```python
#Task 1.7

def quicksort(seq):
    arr = seq
    mid = len(arr) //2
    
    lesser = []
    equal = []
    greater = []

    if len(arr) == 1:
        return arr

    for i in range(len(arr)):
        if arr[i] > arr[mid]:
            greater.append(arr[i])
        elif arr[i] == arr[mid]:
            equal.append(arr[i])
        else:
            lesser.append(arr[i])

    return lesser + equal + greater

bubblesort(lst)
```

```python
#Task 1.8

from timeit import *

insert100 = timeit(lambda: insertionsort(lst), number =1 )
bubble100 = timeit(lambda: bubblesort(lst), number =1 )
quick100 = timeit(lambda: quicksort(lst), number =1 )

print(f"Insertion: \n Order of growth: O(n^2) \n Time complexity:{insert100}\n")
print(f"Bubble: \n Order of growth: O(n^2) \n Time complexity:{bubble100}\n")
print(f"Quick: \n Order of growth: O(nlog(n)) \n Time complexity:{quick100}\n")
```

```python
#Task 2.1



row = ["."] * 7
table = [row] * 10

class board:
    def __init__(self,row,column):
        self.row = row
        self.column = column
        self.table = [['.'] * self.row for i in range(self.column)]

    def viewcoord(self,x,y):
        return(self.table[y][x])

    def updatecoord(self,item,x,y):
        self.table[y][x] = item

    def viewboard(self):
        header = '\ABCDEFGHIJ'
        h = ''  
        print(header)
        for i in range(len(self.table)):
            print(f"{i+1}{h.join(self.table[i])}")
        print("\n")

x = board(10,7)
x.viewboard()
x.updatecoord("h",5,1)
x.viewboard()
```

```python
#Task 2.2

import random


class ship(board):
    def _insertship_h(self):
        valid = False
        while valid == False:
            x = random.randint(0,6)
            y = random.randint(0,6)
            
            for i in range(x,x+4):
                if self.viewcoord(x,y) != '.':
                    valid = False
                    break

            valid = True

        for i in range(x,x+4):
            self.updatecoord("S",i,y)


    def _insertship_v(self):
        valid = False
        while valid == False:
            x = random.randint(0,6)
            y = random.randint(0,3)
            
            for i in range(y,y+4):
                if self.viewcoord(x,i) != '.':
                    valid = False
                    break
            valid = True

        for i in range(y,y+4):
            self.updatecoord("S",x,i)

    def insertship(self):
        x = random.choice([True,False])
        if  x == True:
            self._insertship_h()
        else:
            self._insertship_v()

    def insertrocks(self):
        placed = 0
        while placed <7 :
            x = random.randint(0,9)
            y = random.randint(0,6)
            
            if self.viewcoord(x,y) == '.':
                self.updatecoord('R',x,y)
                placed +=1
            else:
                continue




x = ship(10,7)
x.insertship()
x.insertrocks()

x.viewboard()
```
